## Problem 4 Part 2: Matrix Inverse

Generate 3 random 3×3 matrices with integers from -10 to 10, compute their inverses, and verify A·A⁻¹ = I.

```python
np.random.seed(8)

matrices = [np.random.randint(-10, 11, size=(3, 3)) for i in range(3)]

for i, A in enumerate(matrices):
    print(f"Matrix {i+1}:")
    print(A)
    print("\n")

    det = np.linalg.det(A)
    print(f"Determinant of Matrix {i+1}: {det}")
    print("\n")

    if det == 0:
        print("Matrix is not invertible")
    else:
        inv = np.linalg.inv(A)
        print(f"Inverse of Matrix {i+1}:\n{inv}")
        print("\n")
        print("Verification (A * A^(-1) = I): ")
        print("\n")
        product_I = np.dot(A, inv)
        print(np.round(product_I, 10))
        print("\n")
        print("--------------------------------------------------")

```

## Problem 5: Average, Variance, and Correlation

Exploratory data analysis on the house price dataset.

```python
df = pd.read_csv('kc_house_data.csv')

df = df.drop(columns=['id', 'date', 'zipcode'])

# Making sure the data is loaded correctly
print(f"Dataset shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
df.head()
```

### Part 1: Average, Min, Max, Variance for each feature

```python
statistics = pd.DataFrame({
    'Average': df.mean(),
    # Median because median household statistics are a better measure of central tendency for skewed data and for fun lol
    'Median': df.median(),
    'Mean and Median Difference': np.abs(df.mean() - df.median()),
    'Min': df.min(),
    'Max': df.max(),
    'Variance': df.var()
})

print(statistics.to_string())
print("\n")

lowest_avg_feature = statistics['Average'].idxmin()
highest_avg_feature = statistics['Average'].idxmax()
lowest_var_feature = statistics['Variance'].idxmin()
highest_var_feature = statistics['Variance'].idxmax()

print(f"Feature with Lowest Average: {lowest_avg_feature} ({statistics.loc[lowest_avg_feature, 'Average']:.4f})")
print(f"Feature with Highest Average: {highest_avg_feature} ({statistics.loc[highest_avg_feature, 'Average']:.4f})")
print(f"Feature with Lowest Variance: {lowest_var_feature} ({statistics.loc[lowest_var_feature, 'Variance']:.4f})")
print(f"Feature with Highest Variance: {highest_var_feature} ({statistics.loc[highest_var_feature, 'Variance']:.4f})")
```

### Part 2: Correlation Coefficients with Price

```python
correlations = df.corr()['price'].drop('price')

print("Correlation with price:")
print(correlations.sort_values())
print("\n")

positive_correlations = correlations[correlations > 0]
print(f"Positively correlated features ({len(positive_correlations)}):")
print(list(positive_correlations.index))
print("\n")

print(f"Highest positive correlation: {correlations.idxmax()} ({correlations.max():.4f})")
```

### Part 3: Features with Negative Correlation to Price

```python
negative_correlations = correlations[correlations < 0]

if len(negative_correlations) == 0:
    print("No features have negative correlation with price.")
else:
    print("Features with negative correlation to price:")
    print(negative_correlations.sort_values())
```
